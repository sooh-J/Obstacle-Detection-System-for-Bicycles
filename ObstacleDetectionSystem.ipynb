```python
# close to open
test_files = ["추월2", "도로 1","도로 2",
              "보행자 1","보행자 2",
              "new_자전거_1","new_자전거_2", "new_자전거_3","자전거 1", "자전거 2",
                "new_장애물_1","new_장애물_2"]
mask_runnings = ["mask_running1"]#,"mask_running2"]#,"mask_running3"]
for i,mask_running in enumerate(mask_runnings):
    # print(f"--------------------mask_runnings:{mask_running}--------------------")
    for test_file in test_files:
        print(f"----filename:{test_file}----")

        # preprocessing
        vid_origin, vid_obstacles, vid_running = preprocessing(test_file=test_file, mask_running=mask_running, show=False)

        folder = f"outputs-detection-closetoopen/mask{i}"
        import os
        if not os.path.exists(folder):
            os.makedirs(folder)

        # morphological filtering
        vid_closing, vid_closing_opening = morphological_filtering_closetoopen(vid_running)
        concat_3vid(vid_running,vid_closing,vid_closing_opening,
                    save=True, file_name = f"{folder}/{test_file}_close_close&open")
        # CCL
        vid_ccl_closing, vid_ccl_closing_view = connected_component_labeling(vid_closing_opening)
        concat_3vid(vid_running,vid_closing_opening,vid_ccl_closing_view,
                    save=True, file_name = f"{folder}/{test_file}_mor_mor&ccl-view")
        concat_3vid(vid_running,vid_closing_opening,vid_ccl_closing,
                    save=True, file_name = f"{folder}/{test_file}_mor_mor&ccl")

        # Detection & Classification & draw a box
        print("[[CCL -> OPEN]]")
        org_video, output_video, output_video_color = classification_objects(vid_origin.copy(), vid_ccl_closing.copy())
        concat_3vid(org_video, output_video, output_video_color, save=True, file_name = f"{folder}/{test_file}_detection")
```
